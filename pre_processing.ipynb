```python
import pandas as pd
```

Importing the pandas library

```python
file_path = 'Student Depression Dataset.csv'
df = pd.read_csv(file_path)
```

Loading the dataset from a CSV file into a pandas DataFrame

```python
df_no_id = df.drop(columns=['id'])
```

Creating a new DataFrame without the 'id' column. This column isn't relevant for the model

```python
df_students = df[df['Profession'] == 'Student'].drop(columns=['Profession'])
```

Creating a new DataFrame that includes only rows where the Profession is Student. Then, dropping the Profession column.

```python
mean_value = df_students['Financial Stress'].mean()
df_students['Financial Stress'] = df_students['Financial Stress'].fillna(mean_value)
```

Calculating the mean of the Financial Stress column and filling all the missing values with this value

```python
mapping = {
    'Less than 5 hours': 4.5,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9,
    'Others': None
}
df_students['Sleep Duration'] = df_students['Sleep Duration'].map(mapping)
mean_sleep = df_students['Sleep Duration'].mean()
df_students['Sleep Duration'] = df_students['Sleep Duration'].fillna(mean_sleep)
```

Mapping sleep duration column to numeric values for quantitative analysis. Then, filling missing values with the column's mean

```python
mapping_diet = {
    'Healthy': 3,
    'Moderate': 2,
    'Unhealthy': 1,
    'Others': None
}
df_students['Dietary Habits'] = df_students['Dietary Habits'].map(mapping_diet)
mean_diet = df_students['Dietary Habits'].mean()
df_students['Dietary Habits'] = df_students['Dietary Habits'].fillna(mean_diet)
```

Mapping dietary habit to numeric scores to quantify diet quality. Then, filling missing values with the column's mean

```python
df_students['Lifestyle Score'] = (
    df_students['Sleep Duration'] * 0.4 +
    df_students['Dietary Habits'] * 1.5 -
    df_students['Financial Stress'] * 0.8
)
```

Calculating a new feature Lifestyle Score that reflects overall health by weighting sleep and diet positively, and financial stress negatively

```python
yes_no_mapping = {'Yes': 1, 'No': 0}
df_students['Have you ever had suicidal thoughts ?'] = df_students['Have you ever had suicidal thoughts ?'].map(yes_no_mapping)
df_students['Family History of Mental Illness'] = df_students['Family History of Mental Illness'].map(yes_no_mapping)
```

Mapping Yes/No values in 'Have you ever had suicidal thoughts ?' and 'Family History of Mental Illness' columns to 1 and 0
